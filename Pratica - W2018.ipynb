# Propriedades físicas de nebulosas através de linhas de emissão

## Introdução

Nebulosas planetárias são objetos formados durante os estagios finais da evolução de estrelas de massa intermediária. Estes objetos são bastante importantes pois podem ser usados como ferramentas poderosas em outras áreas da astrofísica. Sua importância vai desde a evolução estelar propriamente dita a evolução de nossa Galáxia como um todo. Como esses objetos são formados pela ejeção de material (enriquecido ou não) da estrela projenitora, eles enriquecem o meio intergalactico.

As características do material ejetado em combinação com o núcleo estelar remanescente faz com que esses objetos emitam linhas espectrais em emissão. Isso se dá pois os núcleos remanescentes são extremamente quentes (da ordem de 100 kK) e portanto emitem radiação de alta energia na faixa do ultra violeta e acima. Essa radiação tem energia suficiente para arrancar elétros dos átomos de hidrogênio e portanto fotoionizar o material ejetado. Quando esses elétros recombinam com o núcleo do atomo de hidrogenio emitindo assim radiação (Figura 1). Algumas das llinhas mais brilhantes e usuais são as llinhas de Balmer, detectadas na região do visível. Veja esse link para mais informações sobre essas linhas: https://en.wikipedia.org/wiki/Balmer_series

Calcule as primeiras 10 linhas de emissão da série de Balmer. Para isso use a equação desenvolvida pelo físico Johannes Rydberg. A equação é dada por:

$${\frac  {1}{\lambda }}=R_{{\mathrm  {H}}}\left({\frac  {1}{2^{2}}}-{\frac  {1}{n^{2}}}\right)\quad {\mathrm  {para~}}n=3,4,5,...$$

onde $\lambda$ é o comprimento de onda e $R_{{\mathrm  {H}}}=1.097 \times 10^7~m^{-1}$ é a constante de Rydberg.
Use o programa a seguir.

```python
# obter o nǘel atomico do usuário
n = input('Entre com o nivel atomico desejado:')
print 'Calculando para n =',n

# calcular o lambda para o respectivo nível
lamb = 1.0 / ( 1.097e7 * (1./2**2 - 1/float(n)**2) )

# imprimir o resultado
print 'O valor de lambda para este nível é: ',lamb*1.0e10, ' angstroms'
```

As linhas espectrais observadas em nebulosas planetárias assim como em outros objetos celestes são de extrema importancia para a astrofísica. Isso porque as linhas espectrais observadas trazem consigo informações sobre os processos físicos que as originaram. Com isso podemos obter informações importantes sobre onde estas linhas se formaram como a densidade, temperatura e composição química.

Dado que as nebulosas contribuem para o enriquecimento químico de galáxias, estudar sua composição é importante. Para obter a abundância química dos elementos presentes no gás da nebulosa planetária precisamos obter as condições físicas deste gás. 

A temperatura e densidade do gás em nebulosas pode ser determinado a partir das linhas espectrais observadas e conhecimento dos processos físicos que formaram essas linhas. Em particular, vamos ver como podemos obter a densidade de um gás fotoionizado.

Íons e elétrons em uma nebulosa planetária estão em constante movimento devido a temperatura em que o gás se encontra. Iso faz com que colisões aconteçam entre estes elétrons e íons. Em geral a energia trocada nessas colisões não é o bastante para arrancar um elétron de um íon mas muitas vezes pode ser suficiente para fazer com que o elétron salte para um nível superior. Ao decair esses elétrons emitem radiação. Essa radiação é observada como linhas espectrais que conhecemos como linhas colisionais devido a maneira com que são formadas. Essas colisões muitas vezes excitam níveis de energia que não seriam excitados de outra forma, devido a sua baixa probabilidade de transição.

Vejamos o caso do enxofre uma vez ionizado por exemplo, que possui 3 elétrons de valência. Sua configuração eletrônica é 1s2 2s2 2p6 3s2 3p3 . Íons com 3p elétrons têm seus níveis mais baixos de energia no estado fundamental. Certas transições de elétrons envolvem níveis de energia metaestáveis; as linhas de emissão resultantes são chamadas de linhas "proibidas" pois níveis metaestáveis tem muito baixa probabilidade de decair radiativamente. A Figura 3 mostra o diagrama de níveis de energia do enxofre uma vez ionizado. Na figura são destacadas duas transições em particular que resultam em linhas espectrais observadas na faixa vísivel do espectro. Essas são as linhas que usaremos para determinar a densidade do gás.

Mas porque podemos usar essas linhas para calcular a densidade? Observe na Figura 3 que as duas linhas destacadas com comprimentos de onda de 6716 e 6731 Angstroms respectivamente são bem proximas em termos de energia. Lembrando que a energia e o comprimento de onda estão relacionados por $E = \frac{hc}{\lambda}$, onde E é energia, h é a constante de Planck , c é a velocidade da luz e λ é o comprimento de onda. Essa proximidade faz com que qualquer colisão nessa faixa de energia possa excitar o eletron para qualquer um dos níveis 5/2 e 3/2.

## Determinando a densidade
(traduzido de https://web.williams.edu/Astronomy/research/PN/nebulae)

Quando a densidade do gás é muito baixa, as colisões são raras, e a maioria dos elétrons que são excitados colisionalmente em níveis superiores são capazes de emitir um fóton quando saltam para o nível de energia inferior. Sob estas condições (abaixo de cerca de $100~cm^{-3}$), a razão entre a linha λ6716 e a linha λ6731 é apenas a razão entre o número de elétrons em cada nível, que é a razão dos pesos estatísticos, que é 6/4=1.5. Os tempos de vida de cada nível não importam aqui, porque mesmo os elétrons no nível 5/2 de vida mais longa têm tempo de irradiar. Quando a densidade do gás é alta e as colisões são muito frequentes, a razão λ6716/λ6731 aproxima-se novamente de um valor constante, desta vez igual à razão entre os pesos estatísticos divididos pela relação entre as vidas úteis. O primeiro fator está lá pelo mesmo motivo que no caso de baixa densidade: quanto mais elétrons um nível puder conter, mais emissões haverá. O segundo fator é necessário, porque agora o tempo de vida é importante: quanto menos tempo um elétron ficar em um nível excitado, mais provável será que ele seja capaz de irradiar antes que outra colisão o derrube, então um tempo de vida curto cria uma linha mais forte. No caso de alta densidade, acima de $10000~cm^{-3}$ , a relação λ6716/λ6731 é de aproximadamente 0,44.

Mas como esses valores constantes são úteis para encontrar a densidade na nebulosa? Eles não são. O que é útil é o seguinte: para densidades entre os extremos descritos acima, a razão de intensidade λ6716/λ6731 observada depende da densidade real. Pense desta forma: à medida que a densidade aumenta a partir de um limite inferior de cerca de $100~cm^{-3}$, o tempo de vida relativa dos dois níveis começa a entrar em jogo à medida que as colisões se tornam mais frequentes. O nível 3/2 de vida mais curta começa a acumular uma pequena vantagem, e a linha λ6731 se fortalece em relação à linha λ6716. À medida que a densidade aumenta, esta tendência continua até cerca de $10000~cm^{-3}$, as colisões são tão frequentes que a vida útil mais curta do nível de 3/2 não confere mais nenhuma vantagem, e a proporção para de aumentar. (Para comparação, o ar que você está respirando agora tem uma densidade de cerca de $10^{19}~cm^{-3}$, ou 10 milhões de bilhões de vezes mais do que uma nebulosa planetária típica!)

A Figura 4 mostra como a relação λ6716/λ6731 se comporta conforme as mudanças de densidade. Observe a  proporção cai suavemente de cerca de 1,5 para 0,44. Outros íons com configurações eletrônicas semelhantes terão seus próprios conjuntos de pesos estatísticos e tempos de vida de nível, e suas proporções serão sensíveis a diferentes densidades.

$$ n_e = \frac{R[SII]-1.49}{5.62 - 12.80 \times R[SII]} \times 10^4$$

## Identificando linhas de emissão em espectros

O que precisamos fazer então é, tendo um espectro de uma nebulosa planetária, identificar as linhas espectrais de interesse e então obter os parametros físicos e químicos do objeto. Para trabalhar com espectros vamos usar o módulo Astropy (http://www.astropy.org/) do Python que possui um ferramental extenso para aplicações em astrofísica.

Os espectros usados em astrofísca, seja de nebulosas ou outros objetos, são essencialmente imagens. Essas imagens podem ser 1D, 2D ou mesmo 3D como no caso de espectroscopia de campo integral(http://astro.if.ufrgs.br/telesc/espectroscopia.html). As imagens astronômicas são armazendas em um formato padrão chamado de FITS. Imagens fits contem os dados binários referente aos fotons coletados efetivamente assim como metadados referente a como estes fotons foram coletados (qual telescópio, tipo de CCD, espectrografo etc). Os metadados são armazenados no que chamamos de cabeçalho da imagem (header em ingles).

Vamos começar então aprendendo a ler uma imagem para podermos analisa-la. Para isso siga o exemplo seguinte que faz a leitura de uma imagem para esta nebulosa feita pelo telescópio Hubble usando um filtro de banda estreita com comprimento de onda central na região das linhas do enxofre uma vez ionizado. ## Determinanndo o fluxo

Agora que temos os dados das linhas de emissão memdidas no espectro, precisamos determinar o fluxo recebido na terra em cada uma dessas transições. Sabemos que as linhas de emissão devm ter uma forma aproximadamente gaussiana (devido a diversos fatores) como pode ser observado nos exercícios anteriores. Para obter o fluxo emitido em uma dada linha assumimos então essa forma e através da integral dessa função obtemos a energia total emitida. A integral de funções gaussianas é bem conhecida (veja em https://pt.wikipedia.org/wiki/Integral_Gaussiana). 

O problema então se torna saber qual é a função gaussiana que representa a linha de emissão que queremos estudar. A função gaussiana é dada por:


$$ f(x)= A~\exp \left(-{\frac {(x+\mu )^{2}}{2\sigma ^{2}}}\right) $$

onde os parametros que precisamos determinar são A, a intensidade da linha, $\mu$ o comprimento de onda central da linha e $\sigma$ a largura da linha.

Para determinar esses parametros da maneira mais robusta e eficiente possível precisamos fazer um ajuste de função em nossos dados observados. Esse procedimento de ajustar funções em dados experimentais é um dos procedimentos científicos mais fundamentais pois é o que nos permite fazer previsões a partir de observações sobre como a natureza funciona. Felizmente existem diversas ferramentas em Python para fazer esse tipo de tarefa. Nessa atividade vamos usar as ferramentas desenvolvidas e disponibilizadas no pacote Astropy: http://docs.astropy.org/en/stable/modeling/ .

Veja no exemplo abaixo o ajuste de um polinomio feito a um conjunto de dados usando as funcionalidades do Astropy.


```python
import numpy as np
from astropy.modeling import models, fitting
from matplotlib import pyplot as plt

# abaixo definimos um conjunto de dados simulados (x,y)
x = np.array([28.3, 32.8 , 33.0, 43.0, 41.1,
       47.4,  2.9, 19.3, 40.8, 12.8,
        3.3, 38.2, 13.7, 20.7, 12.9,
       34.0,  4.1, 28.6,  2.9  , 33.0])

y = np.array([1482.5, 1429.9 , 2260.0, 3789.1,
       4471.5, 5313.9 ,   18.6,  844.0,
       3993.5 ,  394.2,   31.3, 3804.5,
        428.2,  983.9,  191.7, 2033.5,
         44.2, 2552.7,   16.3, 2501.6])


# aqui inicializamos o tipo de modelo que pretendemos ajustar aos nossos dados
# para uma lista de outros modelos veja: 
# http://docs.astropy.org/en/stable/modeling/#module-astropy.modeling.functional_models

p_init = models.Polynomial1D(degree=2)

# define o algoritmo a ser usado para o ajuste
fitter = fitting.LevMarLSQFitter()

# ajusta o modelo escolhido aos dados do problema e salva os resultados na variavel p
p = fitter(p_init, x, y)
print p

# faz um grafico dos dados e do ajuste feito
plt.scatter(x,y)
xplt = np.linspace(0,50,100)
plt.plot(xplt,p(xplt),'r',label='ajuste')
plt.legend()
plt.show()
```
